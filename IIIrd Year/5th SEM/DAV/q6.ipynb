```python


#Consider any sales training/ weather forecasting dataset
#a. Compute mean of a series grouped by another series
#b. Fill an intermittent time series to replace all missing dates with values of previous non-missing date.
#c. Perform appropriate year-month string to dates conversion.
#d. Split a dataset to group by two columns and then sort the aggregated results within the groups.
#e. Split a given dataframe into groups with bin counts.
import pandas as pd
import datetime as dt
import numpy as np
df=pd.read_csv("weatherHistory.csv")
print(df)
df

#a
a=df.groupby("Precip Type")["Humidity"].mean()
a


df.columns
new_df=df.head(10)
new_df

#b
new_df['Formatted Date']=pd.to_datetime(new_df['Formatted Date'],utc=True)
new_df['Formatted Date']
new_df.set_index(new_df['Formatted Date'], inplace=True)
new_df
new_df.asfreq('30T').ffill()



#c
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
df['Formatted Date']
df['new_col']=df['Formatted Date'].dt.strftime('%Y-%m')
print(df['new_col'])
df['new_col']=pd.to_datetime(df['new_col'])
print(df['new_col'])
df['new_col']=df['new_col'].apply(lambda x: dt.datetime(x.year,x.month,10))
df['new_col']



#d
group_df=df.groupby(['Precip Type','Summary'])['Pressure (millibars)'].agg('mean').sort_values(ascending=False)
result=group_df.groupby(level=0,group_keys=False)
result.nlargest()

#e
e=pd.qcut(df["Wind Speed (km/h)"],q=10).value_counts()
e






```
